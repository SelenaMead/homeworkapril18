```python
class Node:
    def __init__(self, value, node=None, prev_node=None):
        self.value = value
        self.next_node = node
        self.previous_node = prev_node
        
    def get_value(self):
        return self.value
    
    def set_value(self, new_value):
        self.value = new_value
        
    def get_next_node(self):
        return self.next_node
    def get_previous_node():
        return self.previous_node
        
    def set_next_node(self, node):
        self.next_node = node
        
    def set_previous_code(self, node):
        self.previous_node = prev_node
        
    def __repr__(self):
        return f'<Node: {self.value}>'

class LinkedList:
    def __init__(self, root=None):
        self.root = root
        self.size = 0
        
    def add(self, number):
        # create the new Node
        new_node = Node(number, self.root)
        
        if self.root:
            self.root.set_previous_node(new_node)
        
        # place it at the very beginning of our LinkedList
        self.root = new_node
        
        # increment the size attribute by +1
        self.size += 1
        
    def remove(self, number):
        # create a starting point with which to create a point of reference to iterate over all
        # our list items
        current_node = self.root
        
        # create a reference to shift the previous node's next_node attribute to the node that came AFTER
        # the node we just removed
        prev_node = None
        next_node = None
        
        while current_node:
            if current_node.get_value() == number:
    
                if prev_node:
                    prev_node.set_next_node(current_node.get_next_node())
                elif next_node:
                    next_node.set_previous_node(current_node.getprevious_node())
                else:
                    self.root = current_node.get_next_node()
                self.size -= 1
                return 'Node removed'
            else:
                prev_node = current_node
                current_node = current_node.get_next_node()
        return 'Node not found'
    
    def find(self, number):
        current_node = self.root
        
        prev_node = None
        
        while current_node:
            if current_node.get_value() == number:
                return True
            else:
                current_node = current_node.get_next_node()
        return False
    
    def get_root(self):
        return self.root
    
    def get_size(self):
        return self.size  
```
